Welcome to New York City, one of the most-visited cities in the world. There are many Airbnb listings in New York City to meet the high demand for temporary lodging for travelers, which can be anywhere between a few nights to many months. In this project, we will take a closer look at the New York Airbnb market by combining data from multiple file types like `.csv`, `.tsv`, and `.xlsx`.

Recall that **CSV**, **TSV**, and **Excel** files are three common formats for storing data. 
Three files containing data on 2023 Airbnb listings are available to you:

**data/airbnb_price.csv**
This is a CSV file containing data on Airbnb listing prices and locations.
- **`listing_id`**: unique identifier of listing
- **`price`**: nightly listing price in USD
- **`nbhood_full`**: name of borough and neighborhood where listing is located

**data/airbnb_room_type.xlsx**
This is an Excel file containing data on Airbnb listing descriptions and room types.
- **`listing_id`**: unique identifier of listing
- **`description`**: listing description
- **`room_type`**: Airbnb has three types of rooms: shared rooms, private rooms, and entire homes/apartments

**data/airbnb_last_review.tsv**
This is a TSV file containing data on Airbnb host names and review dates.
- **`listing_id`**: unique identifier of listing
- **`host_name`**: name of listing host
- **`last_review`**: date when the listing was last reviewed

```python
# We've loaded your first package for you! You can add as many cells as you need.
import numpy as np
import pandas as pd
# Begin coding here ...
```

## Loading the data

```python
# Load CSV
airbnb_price_csv = pd.read_csv('data/airbnb_price.csv')
airbnb_room_type_xlsx = pd.read_excel('data/airbnb_room_type.xlsx')
airbnb_last_review_tsv = pd.read_table('data/airbnb_last_review.tsv')
```

## Merging the three DataFrames

```python
merged = airbnb_price_csv.merge(airbnb_room_type_xlsx, on='listing_id').merge(airbnb_last_review_tsv, on='listing_id') 
merged.head()
```

## Determining the earliest and most recent review dates

```python
merged['last_review_date'] = pd.to_datetime(merged['last_review'], format='%B %d %Y')
earliest = merged['last_review_date'].min()
most_recent = merged['last_review_date'].max()
print(earliest)  

print(most_recent)
```

```python
merged['room_type'].unique()
```

```python
mapping = {'Entire home/apt':'entire home/apt', 'ENTIRE HOME/APT':'entire home/apt', 'Private room':'private room','PRIVATE ROOM':'private room', 'SHARED ROOM':'shared room', 'Shared room':'shared room'}

merged['room_type'] = merged['room_type'].replace(mapping)
merged['room_type'].unique()
```

## Finding how many listings are private rooms

```python
private_rooms = len(merged[merged['room_type'] == "private room"])

print(private_rooms)
```

## Finding the average price of listings

```python
merged['price'] = merged['price'].str.replace("dollars", "").astype(float)
```

```python
average_listing_price = round(merged['price'].mean(), 2) 
average_listing_price
```

## Creating a DataFrame with the four solution values

```python
review_dates = pd.DataFrame({'first_reviewed': earliest, 'last_reviewed':most_recent, 'nb_private_rooms':private_rooms, 'avg_price':average_listing_price}, index=[0])
review_dates
```
